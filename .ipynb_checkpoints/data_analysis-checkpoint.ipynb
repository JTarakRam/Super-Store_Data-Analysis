**Data Description** -

The dataset consists of 51290 rows and 24 columns. Here's a description of each column:<br>
+ Row ID: the unique identifier of each row<br>
+ Order ID: the unique identifier of each order<br>
+ Order Date: the date on which the order was placed<br>
+ Ship Date: the date on which the order was shipped<br>
+ Ship Mode: the shipping mode for the order<br>
+ Customer ID: the unique identifier of each customer<br>
+ Customer Name: the name of the customer who placed the order<br>
+ Segment: the customer segment (Consumer, Corporate, or Home Office)<br>
+ City: the city where the customer is located<br>
+ State: the state where the customer is located<br>
+ Country: the country where the customer is located<br>
+ Postal Code: the postal code where the customer is located<br>
+ Market: the market (US or APAC) where the customer is located<br>
+ Region: the region where the customer is located<br>
+ Product ID: the unique identifier of the product in the order<br>
+ Category: the category of the product (Furniture, Office Supplies, or Technology)<br>
+ Sub-Category: the sub-category of the product<br>
+ Product Name: the name of the product<br>
+ Sales: the sales amount for the product in the order<br>
+ Quantity: the quantity of the product in the order<br>
+ Discount: the discount applied to the product in the order<br>
+ Profit: the profit earned from the product in the order<br>
+ Shipping Cost: the shipping cost for the order<br>
+ Order Priority: the priority level of the order (Low, Medium, High, or Critical)<br>

+ WE have to convert the Order Date, Ship Date into datetime.
+ We have to Divide the data into categorical and numerical and performs our analysis.
+ WE have to remove some of the features which are not usefull for the analysis.

```python
# converting into datetime
data['Order Date'] = pd.to_datetime(data['Order Date'],format= 'mixed')
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'],format= 'mixed')
```

Creating New features based on the order date and ship date, to interpret easily.

```python
# Creating separate columns for year, month, and day
data['Order Year'] = data['Order Date'].dt.year
data['Order Month'] = data['Order Date'].dt.month_name()
data['Order Day'] = data['Order Date'].dt.day_name()
```

```python
# Creating a new column for shipping duration
data['Ship Duration'] = (data['Ship Date'] - data['Order Date']).dt.days
```

```python
# Removing negative values from Ship Duration
data = data[data['Ship Duration'] >= 0]
```

```python
# Creating Profit % feature by the profit and sales
data['Profit %'] = (data['Profit'] / data['Sales']) * 100
```

```python
# separting into numerical and categoical 
num_cols = [col for col in data.columns if data[col].dtype != 'object']
cat_cols = [col for col in data.columns if data[col].dtype == 'object']
```

```python
print(f'Numerical:',num_cols)
print('----------')
print(f'Categorical:',cat_cols)
```

```python
# Num_cols 
num_cols = [col for col in data.columns if data[col].dtype != 'object']
```

```python
# removing outliers using IQR method
for col in num_cols:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    df= data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
```

#### How have monthly sales changed over time?

```python
# Let's see monthly sales in oneyear 
plt.figure(figsize=(16,8))
sns.barplot(data=data, x="Order Month", y="Sales")
plt.show()
```

WE can see that there is no significant difference interms of sales when the month is changed.

```python
# profit in terms of months
plt.figure(figsize=(16,8))
sns.barplot(data=data, x="Order Month", y="Profit")
plt.show()
```

we can see that most of the profit comes in the begining and the ending of the years, where the peoples are tend to buy new products.

```python
# Lets Group the data by month and calculate the total sales for each month
monthly_sales = data.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum()
```

```python
# Creating a line plot of monthly sales
plt.figure(figsize=(16, 8))
plt.plot(monthly_sales.index, monthly_sales.values, color='red')
plt.title('Monthly Sales over Time')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

+ We can clearly see a trend over time, where the month increases the sales are also increasing.
+ There are ups and downs in the trend, there might be some festival seasons where are sales are very high and remaining days are moderate.
